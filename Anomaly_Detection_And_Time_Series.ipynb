Question 1 : What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
- Anomaly Detection - Anomaly detection is the process of identifying observations or data points that deviate significantly from the majority of the data. These unusual points, often called anomalies or outliers, can indicate critical issues such as fraud, system errors, or unusual patterns of behavior. It is widely applied in fields like finance, healthcare, cybersecurity, and manufacturing to monitor and detect irregularities that could have significant consequences.

- Types of Anomalies:
  - Point Anomalies - Point anomalies are single data points that differ significantly from the rest of the data. These anomalies are also known as global outliers.
    - For example, if a customer usually spends ₹50 - ₹100 per day and suddenly makes a ₹ 5000 purchase, this transaction is a point anomaly indicating potential fraud. Similarly, a sudden spike in a temperature sensor from 25°C to 100°C could represent a malfunction.

  - Contextual Anomalies - Contextual anomalies are data points that are considered abnormal only within a specific context or condition. The same value may be normal in one scenario but anomalous in another.
    - For example, a temperature of 32°C may be normal in summer but unusual in winter. Likewise, a sudden spike in web traffic might be expected during a product launch but unusual on a regular day.

  - Collective Anomalies - Collective anomalies occur when a group of related data points is anomalous together, even if individual points appear normal on their own.
    - For example, a sequence of network packets showing unusual communication patterns may indicate a cyberattack. In manufacturing, a combination of slightly abnormal sensor readings, such as vibrations and temperature changes, may collectively signal an impending machine failure.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
- Anomaly detection techniques vary in their methodology and suitability depending on the type of data and anomalies. Three popular methods are Isolation Forest, DBSCAN, and Local Outlier Factor (LOF).

- Isolation Forest - Isolation Forest is a tree-based ensemble method that isolates anomalies instead of profiling normal data. It works by randomly selecting features and splitting values to create partitions; anomalies are isolated faster because they are few and different. This method is effective for high-dimensional datasets and is computationally efficient. Use cases include fraud detection, intrusion detection in cybersecurity, and identifying abnormal sensor readings.

- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)- DBSCAN is a density-based clustering algorithm that groups points closely packed together and identifies points in low-density regions as outliers. It does not require a predefined number of clusters and can handle arbitrarily shaped clusters. Use cases include detecting unusual patterns in geospatial data, network traffic anomalies, and identifying rare events in scientific data. DBSCAN is most effective when anomalies are spatially separated from dense clusters.

- Local Outlier Factor (LOF)- LOF is a distance-based method that evaluates the local density of each data point relative to its neighbors. A point is considered an outlier if its local density is significantly lower than that of its neighbors. LOF is particularly good at detecting contextual anomalies in datasets where the notion of an outlier depends on local structure rather than the global dataset. Use cases include detecting unusual customer behavior in retail, abnormal system logs, and fraud detection in financial transactions.

Question 3: What are the key components of a Time Series? Explain each with one example.
- A time series is a sequence of data points recorded over time, and its behavior is influenced by several core components. Understanding these components helps in analyzing patterns, forecasting future values, and identifying anomalies.

  - Trend- The trend represents the long-term direction of movement in the data—either upward, downward, or stable. It shows how the values of a time series change over an extended period.
    - Example: The monthly sales of a company increasing steadily over five years due to business expansion.

  - Seasonality - Seasonality refers to regular and repeating patterns or fluctuations that occur at fixed intervals (such as daily, weekly, monthly, or yearly).
    - Example: Ice cream sales spike every summer and decrease during winter.

  - Cyclical Component -The cyclical component captures long-term oscillations in the data that occur due to economic or business cycles. Unlike seasonality, cycles do not have a fixed, predictable period.
    - Example: Economic recession and recovery impacting employment rates over several years.

  - Irregular or Random Component - The irregular (random) component consists of unpredictable, short-term variations caused by random or unforeseen events. These do not follow any pattern.
    - Example: A sudden drop in airline passengers due to an unexpected natural disaster.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
- Stationarity in Time Series - A time series is said to be stationary if its statistical properties—such as mean, variance, and autocorrelation—remain constant over time. In other words, a stationary series does not exhibit trends, seasonality, or changing variability. Stationarity is important because many forecasting models (like ARIMA) assume that the underlying data is stable and predictable in its behavior.

- How to Test Stationarity-
  - Plotting the time series helps identify trends, seasonality, or changing variance. A series with constant mean and variance visually indicates stationarity.
  - Dividing the series into different time segments and comparing their means and variances can indicate whether properties remain stable.
  - Augmented Dickey-Fuller (ADF) Test: Checks for the presence of a unit root. A low p-value (typically < 0.05) suggests the series is stationary.
  - KPSS Test: Tests the opposite hypothesis—whether the series is stationary.
  - Phillips-Perron Test: Another test for unit roots.

- How to Transform a Non-Stationary Series into Stationary -
  - Subtracting the current value from its previous value removes trends. First-order differencing usually works; sometimes second-order differencing is needed.
  - Applying a logarithm stabilizes variance, especially in data with exponential growth. Example: Using log(sales) instead of raw sales data.
  - Removing the trend component using regression or moving averages makes the series stationary.
  - Subtracting the value from the same season in the previous cycle helps remove seasonal patterns.
  - A power transformation that helps stabilize variance and make the series more stationary.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
- Autoregressive Model (AR)- The AR model predicts the current value of a time series using a linear combination of its own past values. It assumes that past observations have a direct influence on the future.


  Application: Suitable for data showing correlation with lagged values but without strong seasonality.

  Example: Predicting tomorrow’s stock price using past 5 days' prices (AR(5)).

- Moving Average Model (MA)- The MA model predicts future values using past forecast errors rather than past observations. It models the noise or shocks in the data and how they affect future values.

  Application: Useful when random shocks influence the series, such as error smoothing in financial data.

  Example: MA(2) uses errors from the previous two time steps to predict the next value.

- ARIMA (Autoregressive Integrated Moving Average) - ARIMA combines AR and MA components and includes integration (I), which represents differencing to make a non-stationary series stationary.
Structure: AR (p), differencing (d), MA (q).

  Application: Works well for non-seasonal data with trends but no seasonality.

  Example: Forecasting sales that show upward trend but no repeating seasonal pattern.

- SARIMA (Seasonal ARIMA) - SARIMA extends ARIMA by adding seasonal components for data with repeating patterns. It includes seasonal AR, seasonal differencing, and seasonal MA terms.

  Structure: ARIMA(p, d, q) x (P, D, Q)s where s is seasonal length.
  Application: Ideal for monthly, weekly, or quarterly data that shows clear seasonality.

  Example: Monthly electricity consumption rising every summer.

- SARIMAX (Seasonal ARIMA with Exogenous Variables) - SARIMAX further extends SARIMA by adding exogenous variables (X)—external factors that influence the target series.

  Application: Best for seasonal data where other variables (temperature, marketing spend, holidays, etc.) impact the series.

  Example: Forecasting sales influenced by both seasonal patterns and advertising budget.

```python
'''
Dataset:
● NYC Taxi Fare Data
● AirPassengers Dataset
Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components
'''

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

data = get_rdataset("AirPassengers").data

data['Month'] = pd.to_datetime(data['time'], format='%Y')
data.set_index('Month', inplace=True)

data.rename(columns={"value": "Passengers"}, inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(data['Passengers'], linewidth=2)
plt.title("AirPassengers Time Series", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Passengers", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

decomposition = seasonal_decompose(data['Passengers'], model='multiplicative',
                                   period=12)

fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

```python
'''
Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

np.random.seed(42)

distances = np.random.uniform(0, 20, 500)
fares = distances * 1.8 + np.random.normal(0, 2, 500)

anomaly_dist = np.random.uniform(0, 20, 20)
anomaly_fare = np.random.uniform(50, 150, 20)

distance_all = np.concatenate([distances, anomaly_dist])
fare_all = np.concatenate([fares, anomaly_fare])

data = pd.DataFrame({
    "distance": distance_all,
    "fare": fare_all
})

model = IsolationForest(contamination=0.04, random_state=42)
data["anomaly"] = model.fit_predict(data[["distance", "fare"]])

anomalies = data[data["anomaly"] == -1]
normal = data[data["anomaly"] == 1]


plt.figure(figsize=(10, 6))

plt.scatter(normal["distance"], normal["fare"],
            color="blue", label="Normal", s=40, alpha=0.6)

# Anomalies
plt.scatter(anomalies["distance"], anomalies["fare"],
            color="red", label="Anomaly", s=70, edgecolors="black")

plt.title("Isolation Forest – Anomaly Detection on Taxi Fare Data", fontsize=16)
plt.xlabel("Trip Distance (miles)", fontsize=12)
plt.ylabel("Fare Amount ($)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.

'''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset
from warnings import filterwarnings
filterwarnings("ignore")

data = get_rdataset("AirPassengers").data

data['Month'] = pd.date_range(start="1949-01-01", periods=len(data), freq='M')
data.set_index("Month", inplace=True)

data.rename(columns={"value": "Passengers"}, inplace=True)

ts = data['Passengers']

model = SARIMAX(ts,
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit()

forecast_steps = 12
forecast = results.get_forecast(steps=forecast_steps)

forecast_index = pd.date_range(
    start=ts.index[-1] + pd.offsets.MonthBegin(1),
    periods=forecast_steps,
    freq='M'
)

forecast_series = pd.Series(forecast.predicted_mean.values, index=forecast_index)
conf_int = forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(ts, label="Original", linewidth=2)
plt.plot(forecast_series, label="Forecast (Next 12 Months)", linestyle="--", linewidth=2)

plt.fill_between(
    forecast_index,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.3,
    label="95% Confidence Interval"
)

plt.title("SARIMA Forecast – AirPassengers Dataset", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
'''

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data[:, :2]

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

anomalies = X[y_pred == -1]
normals = X[y_pred == 1]

plt.figure(figsize=(10, 6))

plt.scatter(normals[:, 0], normals[:, 1],
            color='blue', s=40, label='Normal')

plt.scatter(anomalies[:, 0], anomalies[:, 1],
            color='red', s=60, label='Anomalies', edgecolor='black')

plt.title("Local Outlier Factor (LOF) – Anomaly Detection", fontsize=16)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Answers -

- Detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN) - To detect anomalies in 15-minute real-time energy consumption data, I would mainly use Isolation Forest because it is fast, scalable, and well-suited for continuous numerical data streams. It isolates abnormal points using random partitions, making it effective in detecting sudden spikes, drops, or sensor malfunctions in real time. Additionally, Local Outlier Factor (LOF) can be used in parallel when localized anomalies need to be detected, such as when one region behaves unusually compared to neighboring regions. While DBSCAN is a powerful density-based method, it is not ideal for streaming data since it requires the entire dataset and is computationally heavy. Therefore, Isolation Forest is the primary technique, with LOF used for neighborhood-based anomaly refinement.

- Time series model I will use for short-term forecasting (ARIMA / SARIMA / SARIMAX) - For short-term forecasting of 15-minute energy demand, I would use the SARIMAX model because it can capture both seasonality and the effects of external variables. While ARIMA does not handle seasonality and SARIMA does not include exogenous variables, SARIMAX incorporates both seasonal patterns and external factors, making it ideal for energy demand forecasting. This leads to more accurate and reliable predictions, especially when demand changes due to weather events or regional variations.

Logs of detected anomalies would also be stored and reviewed to improve future model performance.

- How would this solution help business decisions or operations - This solution significantly enhances power grid operations by enabling early detection of abnormal consumption patterns, preventing overloads, and reducing the likelihood of outages or equipment failures. Furthermore, real-time anomaly alerts improve decision-making by allowing immediate corrective actions, ensuring stable power delivery and improving customer satisfaction. Overall, the integrated anomaly detection and forecasting system strengthens grid reliability and supports informed business operations.
