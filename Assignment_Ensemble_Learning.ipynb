Question 1: What is Ensemble Learning in machine learning? Explain the key idea behind it.
- Ensemble learning is a machine learning technique that combines multiple individual models (called base learners) to produce a more accurate and robust overall model.

- The key idea behind ensemble learning is that a group of weak learners, when combined properly, can outperform any single strong learner. Each model in the ensemble contributes its prediction, and the final output is usually determined through methods like averaging (for regression) or voting (for classification). This approach helps reduce errors, improves generalization, and makes the model more resilient to overfitting and noise in the data.

---

Question 2: What is the difference between Bagging and Boosting?
- Bagging (Bootstrap Aggregating) and Boosting are both ensemble learning methods, but they differ in how they build and combine models.

  - Bagging trains multiple models independently and in parallel on different random subsets of the training data, created through bootstrap sampling.The final prediction is made by averaging or voting among all models, which helps reduce variance.

  - In contrast, Boosting builds models sequentially, where each new model focuses on correcting the errors made by the previous ones. It assigns higher weights to misclassified samples to improve accuracy.
  
  - While Bagging reduces variance, Boosting reduces bias and often leads to higher accuracy but can be more prone to overfitting.

---

Question 3: What is bootstrap sampling and what role does it play in Bagging methods like Random Forest?
- Bootstrap sampling is a statistical technique that involves randomly selecting samples from the training dataset with replacement, meaning the same instance can appear multiple times in a new dataset.

- In Bagging methods such as Random Forest, bootstrap sampling is used to create diverse training subsets for each individual model or decision tree. This diversity ensures that each model learns slightly different patterns from the data, making the ensemble less sensitive to noise and overfitting. By averaging the results from all models trained on these bootstrap samples, Bagging achieves better stability and accuracy than a single model trained on the full dataset.

---
Question 4: What are Out-of-Bag (OOB) samples and how is OOB score used to evaluate ensemble models?
- Out-of-Bag (OOB) samples refer to the data instances that are not included in a particular bootstrap sample during the Bagging process. Since each model in a Bagging ensemble, such as a Random Forest, is trained on a subset of the data, about one-third of the original data typically remains as OOB samples for that model.

- OOB samples can be used as a validation set to test the model's performance without needing a separate dataset. The OOB score is the average accuracy (or other performance metric) computed using these OOB predictions across all trees in the ensemble. It provides an unbiased estimate of model performance and helps assess generalization ability efficiently.

Question 5: Compare feature importance analysis in a single Decision Tree vs. a Random Forest.
- In a single Decision Tree, feature importance is calculated based on how much each feature contributes to reducing impurity (such as Gini impurity or entropy) at its splitting nodes. However, the importance scores can be biased if the tree overfits or if certain features dominate due to data randomness.

- In a Random Forest, feature importance is more reliable because it is averaged across many trees, each trained on different bootstrap samples and subsets of features. This aggregation reduces bias and variance, leading to a more stable and general representation of which features are truly influential in making predictions.

- Random Forest provides a more robust and trustworthy measure of feature importance compared to a single Decision Tree.

```python
'''
Question 6: Write a Python program to:
● Load the Breast Cancer dataset using
sklearn.datasets.load_breast_cancer()
● Train a Random Forest Classifier
● Print the top 5 most important features based on feature importance scores.
'''

from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
})

# Sort and display the top 5 most important features
top_features = feature_importance_df.sort_values(
    by='Importance', ascending=False).head(5)
print("Top 5 Most Important Features:")
print(top_features)
```

```python
'''
Question 7: Write a Python program to:
● Train a Bagging Classifier using Decision Trees on the Iris dataset
● Evaluate its accuracy and compare with a single Decision Tree

'''

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )

# Decision Tree classifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_pred)

# Bagging Classifier using Decision Trees as base estimators
bagging_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=50,
    random_state=42 )

bagging_model.fit(X_train, y_train)
bagging_pred = bagging_model.predict(X_test)
bagging_accuracy = accuracy_score(y_test, bagging_pred)


print("Accuracy of Single Decision Tree:", dt_accuracy)
print("Accuracy of Bagging Classifier:", bagging_accuracy)
```

```python
'''
Question 8: Write a Python program to:
● Train a Random Forest Classifier
● Tune hyperparameters max_depth and n_estimators using GridSearchCV
● Print the best parameters and final accuracy
'''

from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

# Load the Breast Cancer dataset
data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )


rf_model = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10, 15]
    }

grid_search = GridSearchCV(estimator=rf_model,
                           param_grid=param_grid,
                           cv=5,
                           scoring='accuracy',
                           n_jobs=-1)


grid_search.fit(X_train, y_train)


best_params = grid_search.best_params_


best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
final_accuracy = accuracy_score(y_test, y_pred)

print("Best Hyperparameters:", best_params)
print("Final Accuracy on Test Set:", final_accuracy)
```

```python
'''
Question 9: Write a Python program to:
● Train a Bagging Regressor and a Random Forest Regressor on the California
Housing dataset
● Compare their Mean Squared Errors (MSE)
'''

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

data = fetch_california_housing()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )


bagging_model = BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    n_estimators=100,
    random_state=42
)

bagging_model.fit(X_train, y_train)
bagging_pred = bagging_model.predict(X_test)

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

# MSE
bagging_mse = mean_squared_error(y_test, bagging_pred)
rf_mse = mean_squared_error(y_test, rf_pred)

print("Mean Squared Error (Bagging Regressor):", bagging_mse)
print("Mean Squared Error (Random Forest Regressor):", rf_mse)
```

Answer :

To solve the loan default prediction problem, I decided to apply ensemble learning techniques to improve model performance and reliability. Since predicting default involves both reducing bias and maintaining stability, I carefully compared Bagging and Boosting before finalizing my approach.

- Choose between Bagging or Boosting : I began by evaluating whether Bagging or Boosting would better suit the task. Bagging methods like Random Forest reduce variance and are less prone to overfitting, while Boosting methods such as XGBoost focus on minimizing bias by learning from previous errors. Considering that financial data often has complex, non-linear relationships and sometimes slight imbalance between defaulters and non-defaulters, I chose Boosting because it tends to perform better on such structured datasets and captures subtle patterns effectively.

- Handling overfitting : To avoid overfitting, I took several precautions. First, I used feature selection and removed highly correlated variables to simplify the model. Then, I tuned hyperparameters such as learning rate, maximum tree depth, and number of estimators using cross-validation. I also applied early stopping to prevent the model from training too long and memorizing noise. Regularization parameters like min_child_weight and subsample were also adjusted to make the model more generalizable. Lastly, I monitored the model's performance on a separate validation set to ensure stable results.

- Selecting base models : For the base learner, I used Decision Trees, as they handle both numerical and categorical variables and are easy to interpret. In addition to Boosting, I also trained a Random Forest model as a baseline for comparison. This helped me understand how much additional accuracy Boosting provided over Bagging methods. The Random Forest served as a stable model, while Boosting offered higher predictive accuracy through sequential learning.

- Evaluating performance using cross-validation : To evaluate performance reliably, I used Stratified K-Fold Cross-Validation to maintain the same class distribution across folds. I measured model accuracy, precision, recall, F1-score, and ROC-AUC to ensure balanced performance. Since false negatives (missed defaulters) are more costly in a financial setting, I paid special attention to recall and AUC. After tuning, the Boosting model achieved a higher AUC compared to Random Forest, confirming its superiority for this dataset.

- Justifying ensemble learning in real-world decision-making: Using ensemble learning greatly improved the reliability of loan default predictions. These models reduce errors by combining multiple weak learners, which helps minimize both variance and bias. In a financial institution, this translates to better risk assessment — accurately identifying high-risk borrowers while minimizing false alarms. Moreover, using feature importance and SHAP values from the ensemble helped me explain model predictions to management and compliance teams, ensuring transparency. Overall, ensemble learning provided not just better accuracy, but also more trustworthy and interpretable predictions, leading to smarter lending decisions and reduced default losses.
