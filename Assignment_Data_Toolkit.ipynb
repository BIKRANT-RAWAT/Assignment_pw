1. What is NumPy, and why is it widely used in Python?
- NumPy> NumPy also known as Numerical Python, is a fundamental library for numerical computing. It provides support for large, multi dimensional arrays and matrices, along with a collection of high level mathematical functions to operate on these arrays efficiently.
  It is widely used in Python because:
    * Performance >  NumPy operations are implemented in C and Fortran,    making them significantly faster.
    * Efficiency > It uses contiguous memory blocks for arrays, leading to better cache performance and reduced memory overhead.
    * Foundation for other libraries > Many other popular data science libraries like Pandas, SciPy, Scikit-learn, and Matplotlib are built on top of NumPy arrays.
    * Vectorized operations > It allows for writing concise and efficient code by performing operations on entire arrays without explicit loops.
---

2. How does broadcasting work in NumPy?
- Broadcasting in NumPy is a mechanism that allows NumPy to perform operations on arrays of various shapes. When arrays have different shapes, NumPy automatically stretches or duplicates the smaller array's dimensions to match the larger array's shape that enable element wise operations. This is done without actually creating copies of the data, saving memory and improving performance.
---

6. What are the differences between NumPy arrays and Python lists?
- Difference between NumPy arrays and Python lists are:
    * Data Type >NumPy arrays is Homogeneous (elements must be of the same data type) whereas Python lists is Heterogeneous (elements can be of different data types).
    * Performance >NumPy arrays is Faster for numerical operations due to C implementation and optimized memory layout whereas Python lists is Slower for numerical operations, especially on large datasets.
    * Memory Usage >NumPy arrays is More memory efficient due to fixed data types and contiguous storage whereas Python lists is Less memory efficient due to storing pointers to objects.
    * Functionality >NumPy arrays is Optimized for numerical operations, broadcasting, vectorized operations whereas Python lists is General-purpose, flexible data structure.

8. What does the term "vectorized operation" mean in NumPy?
- Vectorized operation in NumPy means to apply an operation to an entire array or a subset of an array at once, without explicitly writing loop. Instead of iterating through individual elements, the operation is performed on all elements simultaneously, leveraging optimized C implementations for significant performance gains.

17. Explain the difference between apply() and map() in Pandas.
- The difference between apply() and map() in Pandas is:
    * apply() > apply() is a more general purpose function used to apply a function along an axis of a DataFrame or Series. It can operate element-wise, column-wise or row-wise, and is flexible for applying custom functions or built-in functions that are not vectorized in Pandas.
    * map() > map() is specifically designed for element wise transformations on a Pandas Series. It is used to substitute each value in a Series with another value, often based on a dictionary or a function. It's generally faster than apply() for element-wise operations on a Series when a simple mapping is required.

21. Why is NumPy's array slicing faster than Python's list slicing?
- NumPy's array slicing is faster than Python's list slicing because:
    * Contiguous Memory Allocation >  NumPy arrays store elements contiguously in memory, allowing for efficient access and manipulation of data blocks during slicing. Python lists store references to objects, which can be scattered in memory.
    * C Implementation >  NumPy's core operations, including slicing, are implemented in C, which is significantly faster than Python's native operations.
    * No Copying >  When we slice a NumPy array, it often returns a "view" of the original array's data, without creating a new copy in memory, which saves time and memory. List slicing, on the other hand, typically creates a new copy of the sliced portion.
