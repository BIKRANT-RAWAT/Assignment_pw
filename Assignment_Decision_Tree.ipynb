Question 1: What is a Decision Tree, and how does it work in the context of classification?
- A decision tree is a supervised learning algorithm used for both classification and regression tasks. It has a hierarchical tree structure which consists of a root node, branches, internal nodes and leaf nodes. It works like a flowchart help to make decisions step by step where:
  - Internal nodes represent attribute tests
  - Branches represent attribute values
  -  Leaf nodes represent final decisions or predictions.

Working in Classification:

  - Start with the dataset:
We have features (independent variables) and a target class (dependent variable).

  - Select the best feature to split on:
The algorithm chooses the feature that best separates the data into distinct classes.

  - Common criteria:
    - Gini Index (used in CART)
    - Entropy / Information Gain (used in ID3, C4.5)
    - Chi-Square, Gain Ratio, etc.

  - Split the dataset:
Based on the chosen feature, the data is divided into subsets (branches).

  - Repeat recursively:
The process continues on each subset until:
    - All samples in a node belong to the same class, or,
    - There are no remaining features to split on.

  - Prediction:
For a new data point, the model traverses the tree based on feature values until it reaches a leaf node — the class label of that leaf is the prediction.

Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures. How do they impact the splits in a Decision Tree?
- A Decision Tree uses impurity measures like Gini Impurity and Entropy to decide the best places to split the data. These measures help determine how mixed or pure a node is in terms of class distribution. The main idea is that a good split results in child nodes that are as pure as possible — meaning each node ideally contains samples from only one class.
- Gini Impurity :  Gini Impurity measures how often a randomly chosen sample would be incorrectly classified if it were randomly labeled according to the class distribution in the node. It is calculated using the formula:
  - Gini = 1 - ∑(p)^2 , where p is the proportion of samples belonging to class in the node.

  A Gini Impurity of 0 means the node is completely pure (all samples belong to one class), while higher values indicate more impurity. Gini is commonly used in the CART (Classification and Regression Tree) algorithm because it is simple and computationally efficient.

- Entropy as impurity measure :  Entropy, on the other hand, is derived from information theory and measures the level of disorder or uncertainty in a dataset. It is calculated as:
  - Entropy=  -∑ (pi) log2(pi)

  Entropy equals 0 when the node is perfectly pure and reaches its maximum when all classes are equally represented. The Information Gain — the reduction in entropy after a split — helps the tree determine which feature provides the most useful information for classification.


Gini Impurity and Entropy guide the tree in selecting the most informative questions to ask at each node. A feature that produces child nodes with mostly one class (low impurity) is preferred. As the tree continues to split, impurity keeps decreasing until the leaves become pure or no further meaningful splits are possible. Therefore, these impurity measures directly influence the structure, depth, and accuracy of the Decision Tree by ensuring that each split helps the model classify data more effectively.

Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision Trees? Give one practical advantage of using each.
- Pre-pruning and Post-pruning are two techniques used in Decision Trees to prevent overfitting, which occurs when the tree becomes too complex and captures noise instead of useful patterns in the data. Both methods aim to simplify the model and improve its ability to generalize to unseen data, but they differ in when and how the pruning is applied.

  - Pre-pruning (also called early stopping) stops the tree from growing once a certain condition is met. Instead of allowing the tree to expand fully, the algorithm checks after each split whether further division will significantly improve performance. Common stopping criteria include reaching a maximum depth, having a minimum number of samples required to split, or when the information gain from a split is below a certain threshold. This approach saves computation time and prevents the creation of unnecessary branches.
    - Practical Advantage: Pre-pruning is computationally efficient because it avoids building a large and overly complex tree in the first place, making it ideal for large datasets or real-time applications.

  - Post-pruning, also known as cost-complexity pruning or reduced error pruning, allows the Decision Tree to grow to its full depth first and then removes branches that do not contribute significantly to model accuracy. This is usually done by evaluating the performance of subtrees on a validation dataset and pruning those whose removal does not reduce accuracy. Post-pruning tends to produce more accurate models because it evaluates the impact of pruning using actual performance metrics rather than assumptions.
    - Practical Advantage: Post-pruning often results in better generalization and higher accuracy, as it removes only those branches that truly do not help in prediction after assessing their impact on validation data.

Question 4: What is Information Gain in Decision Trees, and why is it important for choosing the best split?
- Information Gain (IG) is a key concept in Decision Trees used to decide which feature and threshold provide the best split at each node. It measures the reduction in uncertainty or impurity after the dataset is split based on a particular attribute. In simple terms, Information Gain tells us how much “information” a feature gives about the target variable.
When building a Decision Tree, the algorithm tries to create branches that make the resulting child nodes as pure as possible — meaning the samples in each node mostly belong to a single class. Information Gain quantifies how much the impurity (measured using Entropy) decreases after the split. The higher the Information Gain, the better the feature is at separating the classes.

- Information Gain is important for choosing the best split in a Decision Tree because it helps the algorithm identify which feature provides the most useful information for classifying the data. The goal of building a Decision Tree is to divide the dataset into smaller, more homogeneous (pure) groups where the target variable (class) is easier to predict.
When a dataset is split based on a feature, the Information Gain measures how much the uncertainty (or impurity) in the target variable decreases after the split. A feature that produces a larger reduction in impurity , meaning it creates purer child nodes ,is considered a better split. Therefore, by calculating the Information Gain for each possible feature and threshold, the algorithm can systematically choose the one that results in the highest gain, ensuring the tree learns meaningful and discriminative patterns.

Question 5: What are some common real-world applications of Decision Trees, and what are their main advantages and limitations?
- Decision Trees are widely used in real-world applications across various industries because they are easy to understand, interpret, and apply to both classification and regression problems. They can handle numerical as well as categorical data, making them highly versatile for practical use.

- In business and finance, Decision Trees are used for credit scoring, loan approval, and fraud detection. Banks and financial institutions use them to decide whether to approve a loan application by evaluating factors such as income, credit history, and employment stability. In healthcare, Decision Trees help in medical diagnosis and disease prediction by identifying key symptoms or test results that indicate a particular illness. They are also used in marketing to predict customer behavior, such as whether a customer will respond to a promotion or churn from a service. In manufacturing, they assist in quality control and fault detection by analyzing production data to identify defective products or process inefficiencies.

- Advantages :  The main advantages of Decision Trees are their simplicity, interpretability, and flexibility. They are easy to visualize and explain, even to non-technical users, which makes them useful for decision-making in business contexts. Decision Trees require little data preprocessing — no need for feature scaling or normalization — and can handle non-linear relationships between features. They also work well with both small and large datasets.

- Disadvantage :   Decision Trees also have limitations. One major drawback is their tendency to overfit, especially when the tree grows too deep and captures noise from the training data. They are also sensitive to small changes in the data — a slight variation can lead to a completely different tree structure. Additionally, Decision Trees can become biased if some classes dominate the dataset, and they may not always perform as well as ensemble methods like Random Forests or Gradient Boosted Trees.

```python
'''
Question 6: Write a Python program to:
● Load the Iris Dataset
● Train a Decision Tree Classifier using the Gini criterion
● Print the model’s accuracy and feature importances
(Include your Python code and output in the code box below.)

'''

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from warnings import filterwarnings
filterwarnings('ignore')

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )

clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Decision Tree Classifier using Gini Index\n")
print(f"Model Accuracy: {accuracy:.2f}")
print("\nFeature Importances:")
for name, importance in zip(iris.feature_names, clf.feature_importances_):
    print(f"{name}: {importance:.3f}")
```

```python
'''
Question 7: Write a Python program to:
● Load the Iris Dataset
● Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to
a fully-grown tree
'''

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )

full_tree = DecisionTreeClassifier(criterion='gini', random_state=4)
full_tree.fit(X_train, y_train)

limited_tree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=2)
limited_tree.fit(X_train, y_train)

y_pred_full = full_tree.predict(X_test)
y_pred_limited = limited_tree.predict(X_test)

accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_limited = accuracy_score(y_test, y_pred_limited)

print(f"Accuracy of Fully-Grown Tree: {accuracy_full:.2f}")
print(f"Accuracy of Tree with max_depth=3: {accuracy_limited:.2f}")

```

```python
'''
Question 9: Write a Python program to:
● Load the Iris Dataset
● Tune the Decision Tree’s max_depth and min_samples_split using
GridSearchCV
● Print the best parameters and the resulting model accuracy
'''

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
    )

dt = DecisionTreeClassifier(random_state=42)

param_grid = {
    'max_depth': [2, 3, 4, 5, 6, None],
    'min_samples_split': [2, 3, 4, 5, 6, 10]
}

grid_search = GridSearchCV(
    estimator=dt, param_grid=param_grid, cv=5, scoring='accuracy'
    )

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Decision Tree Hyperparameter Tuning (Iris Dataset)\n")
print("Best Parameters Found:" , best_params)
print(f"\nModel Accuracy with Tuned Parameters: {accuracy:.2f}")
```

- Handle the missing values:  First, identify the type and pattern of missingness (MCAR, MAR, MNAR) and how much is missing per feature. For small amounts of random missingness, simple imputation works: use median for numeric features (robust to outliers) and most frequent for categorical features. For larger or more structured missingness, prefer model-based imputations (KNN or iterative/MICE) or using learned imputers that respect correlations. Always consider adding a missing indicator column (is_missing) for features where “missing” itself may be informative (common in healthcare). Do imputation inside a pipeline (not before splitting) to avoid leakage — fit imputers only on training data.

- Encode the categorical features :  Separate categorical variables by cardinality. For low-cardinality nominal features, use One-Hot Encoding (using OneHotEncoder(handle_unknown='ignore')) inside the pipeline so the model handles unseen categories at inference. For high-cardinality features (e.g., thousands of diagnosis codes), use target / mean encoding or frequency encoding — but compute them using cross-validation folds or smoothing to avoid leakage/overfitting. If any categorical features are ordinal (natural order), map them to integers. When using Decision Trees, encoding doesn’t need scaling, but consistent handling of unknowns and missing values is crucial.

- Train a Decision Tree model :  Split data with train_test_split to preserve class ratios. If classes are imbalanced (likely in disease prediction), either: (a) use class_weight='balanced' in DecisionTreeClassifier, (b) resample the training set with SMOTE/ADASYN for minority upsampling or random undersampling for majority, or (c) combine both carefully. Build a reproducible pipeline with ColumnTransformer for preprocessing (imputers + encoders) and the DecisionTreeClassifier as the final estimator. Use stratified cross-validation when estimating performance to get reliable metrics.

- Tune its hyperparameters : Using GridSearchCV or RandomizedSearchCV wrapped around the full pipeline to tune hyperparameters, this ensures preprocessing is fitted per fold and prevents leakage. Key Decision Tree hyperparameters to tune:
  - max_depth (controls complexity)
  - min_samples_split, min_samples_leaf
  - max_features (features considered per split)
  - criterion (gini or entropy)
  - ccp_alpha (cost-complexity pruning parameter)

  For imbalanced problems tune with scoring set to a metric aligned with business needs (e.g., f1, roc_auc, or average_precision for rare diseases). Use stratified CV (cv=StratifiedKFold) and consider RandomizedSearchCV first if the grid is large, then refine with a grid around promising values.

- Evaluate its performance: Evaluate using appropriate metrics beyond accuracy, Recall (sensitivity) is often crucial in healthcare (catch as many true disease cases as possible), Precision (to avoid too many false positives), F1-score, ROC AUC, and PR AUC (especially for imbalanced classes). Produce a confusion matrix, calibration curve (how well predicted probabilities match observed risk), and decision-curve analysis if available (clinical utility). Validate the final tuned model on a held-out test set and, if possible, on an external dataset (temporal or geographic holdout). Also check fairness/performance across subgroups (age, sex, ethnicity) and inspect feature importances and local explanations (SHAP) so clinicians can understand model reasoning. Finally, plan for model monitoring in production: data drift alerts, periodic re-evaluation, and logging of predictions vs outcomes.

- Business value in the real world :  A well-built disease-prediction Decision Tree can deliver immediate business and clinical value: early detection and risk stratification (triage patients for further testing), prioritization of scarce resources (e.g., specialist referrals, diagnostic tests), reduction of downstream costs through earlier intervention, and improved patient outcomes through targeted screening. Because trees are inherently interpretable, they are easier to validate with clinicians and to use as part of decision support systems in Electronic Health Records (EHRs). Additionally, the model can identify actionable risk drivers (which features most influence predictions), guiding preventive measures or protocol changes.
