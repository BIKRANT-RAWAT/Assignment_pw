Question 1: What is the role of filters and feature maps in Convolutional Neural Network (CNN)?
- In a Convolutional Neural Network (CNN), filters (also called kernels) and feature maps play a fundamental role in enabling the network to automatically learn and extract meaningful patterns from input data such as images. Filters are small matrices of learnable weights that slide (convolve) over the input image or the output of a previous layer. As a filter moves across the input, it performs element-wise multiplication and summation with the local region it covers, producing a single value at each position. Each filter is designed to detect a specific type of feature, such as edges, corners, textures, or more complex patterns. During training, the values of these filters are optimized so that they respond strongly to important visual patterns relevant to the task.

- The output generated after applying a filter across the entire input is called a feature map. A feature map highlights the locations where the corresponding filter has detected its specific feature. For example, an edge-detecting filter will produce high values in regions where edges are present in the image and low values elsewhere. Multiple filters are used in a single convolutional layer, and each filter produces its own feature map, allowing the CNN to capture different types of features simultaneously. This parallel extraction of features makes CNNs highly effective in processing visual data.

- As CNNs become deeper, early layers learn low-level features such as edges and color gradients, while deeper layers learn high-level and abstract features like shapes, objects, or facial structures by combining information from multiple feature maps. Feature maps also help reduce the number of parameters compared to fully connected networks by using weight sharing, which improves computational efficiency and reduces overfitting.

Question 2: Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?

- Padding and stride are two important hyperparameters that control how the convolution operation is applied and how the spatial dimensions of feature maps change across layers.
  - Padding refers to the practice of adding extra pixels, usually zeros, around the border of the input image before applying the convolution operation. The main purpose of padding is to control the size of the output feature map and to ensure that important information at the edges of the image is not lost. Without padding, the convolution operation reduces the spatial dimensions of the input, causing border pixels to be used fewer times than central pixels. By using appropriate padding, such as same padding, the output feature map can retain the same height and width as the input, while valid padding applies no padding and results in a smaller output.
  
  - Stride defines the number of pixels by which the filter moves across the input image during convolution. A stride of 1 means the filter moves one pixel at a time, producing a dense and detailed feature map. Increasing the stride value causes the filter to move more pixels at each step, which reduces the spatial dimensions of the output feature map. This leads to downsampling, lowering computational cost and making the model more efficient, but it may also result in loss of fine-grained spatial details.

The combined effect of padding and stride directly determines the output dimensions of the feature map. Mathematically, for an input of size NxN, filter size FxF, padding P, and stride S, the output dimension is given by

Output size = ( (N-F+2P)/S ) +1

Padding increases the output size by adding border pixels, while stride reduces the output size by skipping positions during convolution.

Question 3: Define receptive field in the context of CNNs. Why is it important for deep architectures?
- The receptive field refers to the specific region of the input image that influences the activation of a particular neuron (or unit) in a feature map. In other words, it defines how much of the original input a neuron can “see” when computing its output. In the first convolutional layer, the receptive field is equal to the size of the convolutional filter (for example, a 3x3 filter sees only a small local region of the image). However, as we move deeper into the network, the receptive field of neurons increases because each neuron in higher layers is connected to a combination of activations from the previous layers. As a result, deeper neurons can capture information from larger portions of the input image.

- The concept of receptive field is crucial for deep architectures because visual understanding requires both local and global context. Early layers with small receptive fields are effective at learning low-level features such as edges, textures, and simple patterns. As the network becomes deeper, the increasing receptive field allows neurons to integrate these low-level features into higher-level representations such as shapes, object parts, and entire objects. This hierarchical growth of receptive fields enables CNNs to model complex spatial relationships and long-range dependencies in images.

  Moreover, deep architectures allow receptive fields to grow efficiently without using very large filters, which would significantly increase computational cost and parameters. By stacking multiple convolutional layers with small filters, CNNs achieve large effective receptive fields while maintaining parameter efficiency and better generalization. Therefore, the receptive field is important because it determines the scale of information captured at each layer, and in deep CNNs, it enables the network to progressively learn from fine-grained local details to broad, high-level semantic understanding of the input.

Question 4: Discuss how filter size and stride influence the number of parameters in a CNN.
- In a Convolutional Neural Network (CNN), filter size and stride play an important role in determining the number of parameters and the overall complexity of the network. The filter size directly affects the number of learnable parameters in a convolutional layer. For a filter of size FxF applied to an input with C channels and using K filters, the total number of parameters is Kx(FxFxC+1), including bias. Larger filters therefore have more weights, increasing memory requirements and the risk of overfitting, while smaller filters such as 3x3 are more parameter-efficient and are widely used in deep CNN architectures to capture complex patterns by stacking multiple layers.

- Stride, in contrast, does not directly affect the number of parameters because it does not change the size or number of filters. However, stride controls how the filter moves across the input and thus determines the spatial dimensions of the output feature maps. A larger stride results in smaller feature maps, reducing the number of activations passed to subsequent layers. This indirectly lowers the number of parameters in later layers, especially fully connected layers.

Question 5: Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.

- LeNet, AlexNet, and VGG are classic CNN-based architectures that represent the evolution of deep learning models for image recognition, differing mainly in depth, filter sizes, and performance.

  - LeNet, one of the earliest CNN architectures, is relatively shallow, consisting of about 5-7 layers including convolutional and fully connected layers. It uses small input images (such as 32x32) and relatively simple filter sizes, making it computationally efficient but limited in representational power. As a result, LeNet performs well on simple tasks like handwritten digit recognition (MNIST) but struggles with complex, large-scale image datasets.
  
  - AlexNet marked a major breakthrough in deep learning by significantly increasing network depth to around 8 layers and introducing larger filter sizes, such as 11x11 and 5x5, in the initial layers. It was designed to process larger images and demonstrated superior performance on the ImageNet challenge, largely due to its ability to learn richer features. AlexNet also introduced key innovations such as ReLU activation, dropout for regularization, and GPU-based training, which greatly improved training speed and generalization compared to LeNet.
  
  - VGG further increased depth, using 16 or 19 layers, but unlike AlexNet, it relied exclusively on small 3x3 filters stacked across many layers. This design choice allowed VGG to achieve large effective receptive fields while keeping the architecture simple and uniform. The increased depth enabled VGG to learn highly detailed and hierarchical features, leading to improved accuracy over AlexNet on large-scale image recognition tasks. However, this came at the cost of a very large number of parameters and high computational requirements.

```python
'''
Question 6: Using keras, build and train a simple CNN model on the MNIST dataset
from scratch. Include code for module creation, compilation, training, and evaluation.

'''

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.utils import to_categorical

# Load dataset
(X_train, y_train), (X_test, y_test) = datasets.mnist.load_data()

# Reshape and normalize
X_train = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train, y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.1
)

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
'''
Question 7: Load and preprocess the CIFAR-10 dataset using Keras, and create a
CNN model to classify RGB images. Show your preprocessing and architecture.
'''

import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.utils import to_categorical

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1
)

test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
'''
Question 8: Using PyTorch, write a script to define and train a CNN on the MNIST
dataset. Include model definition, data loaders, training loop, and accuracy evaluation
'''

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_dataset = datasets.MNIST(
    root='./data', train=True, download=True, transform=transform
)
test_dataset = datasets.MNIST(
    root='./data', train=False, download=True, transform=transform
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 12 * 12, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CNN().to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 5
for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader):.4f}")

model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```
