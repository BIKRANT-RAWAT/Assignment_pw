Question 1: What is hypothesis testing in statistics?
- Hypothesis testing is a statistical procedure used to make inferences about a population parameter based on sample data. It provides a systematic method to decide whether there is sufficient evidence to reject a stated assumption.
Steps involved:
  - Formulation of hypotheses:
    - Null hypothesis (H₀): Represents no effect or no difference
    - Alternative hypothesis (H₁/Ha): Represents the claim to be tested
    - Selection of significance level (α): Usually 0.05 or 0.01, it represents the probability of making a Type I error.
    - Choice of test statistic: Depending on data type and sample size (e.g., z-test, t-test, chi-square test).
    - Decision rule: Compare the calculated test statistic or p-value with the critical value.
    - Conclusion: Reject H₀ if evidence is strong; otherwise, fail to reject H₀.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
- In hypothesis testing we have two types null hypothesis and alternate hypothesis.
  - Null Hypothesis (H₀): The null hypothesis is a statistical statement that assumes no effect, no difference, or no relationship exists in the population. It represents the “status quo” or baseline condition that researchers attempt to test against. Example: H₀: The mean exam score of students = 50.

  - Alternative Hypothesis (H₁ or Ha): The alternative hypothesis is the statement that contradicts the null hypothesis. It represents the researcher's claim or the effect/difference they expect to find Example: H₁: The mean exam score of students ≠ 50.
  
Key differences between null hypothesis and alternative hypothesis are:
  - The null hypothesis (H₀) assumes no effect, no difference, or no relationship, whereas the alternative hypothesis (H₁/Ha) assumes the presence of an effect, difference, or relationship.
  - H₀ represents the default or baseline assumption, while H₁ represents the researcher's claim to be tested.
  - H₀ is tested for possible rejection, while H₁ is considered if H₀ is rejected.
  - H₀ is usually expressed with “=” (equal to), while H₁ is expressed with “≠, <, or >” depending on the test.
  - Example: H₀: μ = 50 (no difference), H₁: μ ≠ 50 / μ > 50 / μ < 50 (difference exists).

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- Significance Level in Hypothesis Testing : The significance level (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true. In other words, it represents the risk of making a Type I error. Frequently chosen values are 0.05 (5%), 0.01 (1%), or 0.10 (10%). For example, α = 0.05 means there is a 5% chance of wrongly rejecting H₀.

Role in Decision-Making: After calculating the p-value or test statistic, the significance level is used as the cut-off. If p-value ≤ α, we reject H₀ and conclude there is enough evidence to support H₁. If p-value > α, we fail to reject H₀, meaning there is not enough evidence against it. It controls the probability of drawing incorrect conclusions.

Question 4: What are Type I and Type II errors? Give examples of each.
- Type I and Type II Errors in Hypothesis Testing :

    - Type I Error (α error): It occurs when the null hypothesis (H₀) is true but is wrongly rejected. In other words, we conclude that an effect or difference exists when it actually does not. The probability of making a Type I error is equal to the significance level (α). Example: A medical test indicates that a healthy patient has a disease (false positive).
    
    - Type II Error (β error): It occurs when the null hypothesis (H₀) is false but we fail to reject it. In other words, we conclude that there is no effect or difference when one actually exists. The probability of making a Type II error is denoted by β, and (1 - β) is called the power of the test. Example: A medical test fails to detect a disease in a sick patient (false negative).

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
- Difference between Z-test and T-test are :

    - Z-test: A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two samples) when the population variance is known and the sample size is large (n > 30). It is based on the standard normal distribution (Z-distribution).
    Example: Testing whether the average height of 1000 students equals 160 cm, when population variance is known.
    
    - T-test: A T-test is used when the population variance is unknown and must be estimated from the sample data. It is mainly applied when the sample size is small (n ≤ 30). It follows the Student's t-distribution, which is wider and accounts for extra uncertainty due to smaller samples.
    Example: Testing whether the mean marks of 20 students in a class differ from 50 when variance is unknown.

```python
'''
Question 7: Implement hypothesis testing using Z-statistics for a
sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)
'''
import numpy as np
import scipy.stats as stats

#given sample data in ques above
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
# done ddof=1 for sample standard deviation
n = len(sample_data)

# Z-statistic calculation using z test formula
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Sample Size:", n)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation at 5% significance level that is alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print('Conclusion :Reject the null hypothesis.')
else:
    print("Conclusion: Fail to reject the null hypothesis.")
```

```python
'''
Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores
represent in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)
'''

import numpy as np
import matplotlib.pyplot as plt

#calculating Z-scores
def calculate_zscores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std deviation
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Sample dataset creating
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5

# calling Z-scores to calc

z_scores = calculate_zscores(data)

# Ploting histogram of standardized data
plt.hist(z_scores, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(0, color="red", linestyle="dashed", linewidth=1.5, label="Mean (0)")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

- What Z-scores represent: A Z-score tells how many standard deviations a value is away from the mean. Example:
  - Z = 0 → the value is exactly at the mean.
  - Z = +1 → the value is 1 standard deviation above the mean.
  - Z = -2 → the value is 2 standard deviations below the mean.

Since Z-scores are standardized, they are centered around 0 (mean).
Most values fall between -2 and +2 (68 - 95 - 99.7 rule for normal distributions)
